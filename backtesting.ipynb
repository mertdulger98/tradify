```python
import pandas as pd
import numpy as np
import yfinance as yf
from ta_backtesting import PairsTrading
```

```python
pt = PairsTrading(
    start="2022-01-01",
    end="2024-05-01",
    approach="log",
    window=10,
    zscore_treshold=1,
    timeframe = '1d'
    )
```

```python
df = pt.backtest(("AAPL","TSLA"))
df
```

```python
df.set_index(df.columns[0])['cumulative_return'].plot()
```

```python
df.set_index(df.columns[0])[list(df.columns)[-3:]].plot()
```

```python
df.set_index(df.columns[0])[list(df.columns)[-5:]].plot()
```
