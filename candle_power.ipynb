```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
df = yf.download("QQQ",period="5y",interval="1d").reset_index()
```

```python
df['rate']=((df['Close']/df['Open'])-1)*100
```

```python
df['body-wick'] = np.abs(df['Close']-df['Open'])/np.abs(df['High']-df['Low'])
```

```python
df['upshadow']=np.where((df['Close']/df['Open'])-1>0,
                       np.abs(df['High']-df['Close'])/np.abs(df['High']-df['Low']),
                       np.abs(df['High']-df['Open'])/np.abs(df['High']-df['Low']))
```

```python
df['downshadow']=np.where((df['Close']/df['Open'])-1>0,
                       np.abs(df['Low']-df['Open'])/np.abs(df['Low']-df['High']),
                       np.abs(df['Low']-df['Close'])/np.abs(df['Low']-df['High']))
```

```python
df['bullpwr']=df['body-wick']+df['downshadow']
df['bearpwr']=df['body-wick']+df['upshadow']
```
