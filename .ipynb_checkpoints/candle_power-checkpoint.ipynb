```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = yf.download("QQQ",period="5y",interval="1d").reset_index()
```

```python
df['rate']=((df['Close']/df['Open'])-1)*100
```

```python
df['body-wick'] = np.abs(df['Close']-df['Open'])/np.abs(df['High']-df['Low'])
df['body-wick2'] = (df['Close']-df['Open'])/(df['High']-df['Low'])
```

```python
df['upshadow']=np.where((df['Close']/df['Open'])-1>0,
                       np.abs(df['High']-df['Close'])/np.abs(df['High']-df['Low']),
                       np.abs(df['High']-df['Open'])/np.abs(df['High']-df['Low']))
```

```python
df['downshadow']=np.where((df['Close']/df['Open'])-1>0,
                       np.abs(df['Low']-df['Open'])/np.abs(df['Low']-df['High']),
                       np.abs(df['Low']-df['Close'])/np.abs(df['Low']-df['High']))
```

```python
df['magnitude']=np.abs(df['rate'])
```

```python
df['r3']=df['rate'].rolling(3).sum().shift(-3)
```

```python
df['color']=np.where(df['rate']>0,'Y','K')
```

```python
df = df[['Date','rate','body-wick2','upshadow','downshadow','color','r3']]
```

```python
df['updown']=df['upshadow']/df['downshadow']
df['downup']=df['downshadow']/df['upshadow']
df['shadow']=df['downshadow']-df['upshadow']
```

```python
df['power']=df['body-wick2']+df['shadow']
df['power2']=df['body-wick2']+df['shadow']*0.8
```

```python
df['pwr_mean']=df['power2'].rolling(3).mean().shift(1)
df['pwr_sum']=df['power2'].rolling(3).sum().shift(1)
```
