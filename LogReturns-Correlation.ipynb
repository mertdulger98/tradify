```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
def getData(stockName, period='5y', interval='1d'):
    return yf.download(tickers=stockName, period=period, interval=interval)
```

```python
def calcReturn(df,ticker):
    df[f'Return_{ticker}']=np.log(df['Close'] / df['Close'].shift(1))
    return df[f'Return_{ticker}'].to_frame()
```

```python
dfs_to_concat = []
for i in stc:
    st = calcReturn(getData(i),i)
    dfs_to_concat.append(st)
    
df = pd.concat(dfs_to_concat, axis=1)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
corr = df.corr()
linked = linkage(corr, 'ward')
# Plot the dendrogram
plt.figure(figsize=(13, 12))
dendrogram(linked, labels=corr.columns, orientation='top')
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Stocks")
plt.ylabel("Euclidean Distance")
plt.show()
```

A dendrogram, as used in hierarchical clustering, is a tree diagram used to visualize the arrangements of clusters formed by hierarchical clustering. The x-axis represents the observations (in this case, the stocks), and the y-axis represents the distance or dissimilarity between these observations. Here's how to interpret the plot:

Leaves: The individual labels at the bottom of the dendrogram are called leaves, and they represent the stocks you're observing (e.g., AAPL, GOOGL, etc.).

Branches: The lines connecting the leaves or other branches are called branches. A set of leaves connected by a single branch indicates that those stocks are similar to each other, based on their returns.

Height of Branches: The vertical lines represent the clusters being merged. The height of the vertical line (y-axis value) represents the distance (dissimilarity) at which they were merged. A higher vertical line indicates that the clusters being merged are more dissimilar compared to clusters merged by a lower line.

Distance or Dissimilarity: In this context, the distance between two clusters is based on their correlation. It indicates how different the stock returns are from each other. You've used Euclidean distance based on the correlation matrix in the example.

Number of Clusters: You can determine the optimal number of clusters by cutting the dendrogram at a certain height and counting the number of vertical lines it intersects. For example, if you cut the dendrogram at a height where it intersects two lines, it means that your data can be optimally grouped into two clusters.

To explain the dendrogram:

Identify the stocks that are closely related: These are stocks whose branches merge at a lower height, indicating high similarity (or correlation) in their returns.
Identify outliers or dissimilar stocks: These are stocks whose branches merge at a higher height, indicating low similarity (or correlation) in their returns.
Determine optimal number of clusters: By observing at what height most of the branches merge, you can estimate the optimal number of clusters that represent the data well.
For instance, if you see that AAPL and MSFT merge at a very low height, it suggests that their stock returns are very similar. Conversely, if TSLA merges with the rest of the stocks at a much higher height, it suggests that TSLA's returns are quite distinct compared to the others.

```python
# Define linkage methods
methods = ['single', 'complete', 'average', 'ward']

# Plot dendrograms for different linkage methods
plt.figure(figsize=(20, 10))

for i, method in enumerate(methods, 1):
    plt.subplot(2, 2, i)
    
    # Compute linkage matrix
    linked = linkage(corr, method=method)
    
    # Plot dendrogram
    dendrogram(linked,labels=corr.columns)
    
    plt.title(f'Hierarchical Clustering with {method.capitalize()} Linkage')
    plt.xlabel('Sample Index')
    plt.ylabel('Distance')

plt.tight_layout()
plt.show()
```

By comparing the dendrograms, you can observe how the choice of linkage method affects the clustering results. 
For instance, "single" linkage might produce longer and more skewed dendrograms, 
while "ward" usually generates more balanced dendrograms.







Single Linkage (Minimum Linkage):

Description: Considers the shortest distance between all pairs of points, one from each cluster, as the distance between the two clusters.
Characteristic: Tends to produce elongated or chain-like clusters. It is sensitive to outliers.


Complete Linkage (Maximum Linkage):
Description: Considers the longest distance between all pairs of points, one from each cluster, as the distance between the two clusters.
Characteristic: Tends to produce compact and roughly equally sized clusters. It is less sensitive to outliers.


Average Linkage:
Description: Considers the average distance between all pairs of points, one from each cluster, as the distance between the two clusters.
Characteristic: Tends to produce clusters that are a compromise between single and complete linkage. It is less sensitive to noise and outliers.


Ward's Method:
Description: Considers the increase in the sum of squared distances within all clusters when two clusters are merged. It aims to minimize the increase in total within-cluster variance after merging.
Characteristic: Tends to create compact and equally sized clusters. It is often more computationally intensive than the other methods.
