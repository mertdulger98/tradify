```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
def getData(stockName,period='60d',interval='15m'):
    return yf.download(tickers=stockName, period=period, interval=interval)['Close'].rename(stockName).to_frame()
```

```python
xl = ['kchol','thyao','bimas','tuprs','sasa','sise','akbnk','froto','tcell','eregl']
xl = [i.upper()+'.IS' for i in xl]
xd = [getData(i) for i in xl]
```

```python
df=pd.concat(xd,axis=1)
```

```python
df[df.index<'2023-06-09']
```

```python
round(df.corr(),2)
```
